```python
#Partners with Swathi Arvind
#Load the dataset from ‘train.csv’ into a pandas DataFrame
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
df = pd.read_csv("train.csv")

df
```

```python
#and print it along with its shape.
shape = df.shape
print('\nDataFrame Shape :', shape)
```

```python
#For dealing with missing values, delete a column with any Null value
#Also dropped ID column
df = df.dropna(axis=1)
df.drop('Id', inplace=True, axis=1)
```

```python
df
```

```python
#For simplicity, we will work with all the numerical columns but just a subset of the
#categorical columns. Among all the remaining categorical columns after step 2,
#choose the categorical columns that have less than 10 unique values.
for i in df.columns:
    if(df[i].dtype)=='object':
        if df[i].nunique() >=10:
          df.drop(columns=i, inplace=True)
```

```python
df
```

```python
#Divide the data into X and Y, where X is the set of features and Y is the target
#For step 5, two approaches will be used (as discussed in class): categorical features.
target = df['SalePrice']
#a. Set of ONLY numerical features and dropping all Categorical features
numericalOnly = df.copy(deep=True)
#b. Set of features comprising both numerical and one-hot encoded
oneHot = df.copy(deep=True)
for i in numericalOnly.columns:
    if(numericalOnly[i].dtype)=='object':
      numericalOnly.drop(columns=i, inplace=True)
numericalOnly.drop(columns='SalePrice', inplace=True)
numericalOnly
```

```python
#Print the original dataframe joined with the one-hot encoded columns.
oneHot.drop(columns='SalePrice', inplace=True)
oneHot = pd.get_dummies(oneHot)
oneHot
```

```python
#Once, the two sets of features are obtained from step 6, train a simple Linear
#Regression model from sklearn and obtain the Mean Absolute Error for both the
#cases:
#a. Mean Absolute Error when Dropping Categoricals:
#b. Mean Absolute Error with One-Hot Encoding
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
regressor = LinearRegression()
X_trainOH, X_testOH, y_trainOH, y_testOH = train_test_split(oneHot, target, test_size=0.5, random_state=3)
X_trainN, X_testN, y_trainN, y_testN = train_test_split(numericalOnly, target, test_size=0.5, random_state=3)
regOneHot = LinearRegression().fit(X_trainOH, y_trainOH)
regNumerical = LinearRegression().fit(X_testN, y_trainN)
y_predOH = regOneHot.predict(X_testOH)
y_predN = regNumerical.predict(X_testN)
print("MAE for ONEHOT", mean_absolute_error(y_testOH, y_predOH))
print("MAE for Numbers Only", mean_absolute_error(y_testN, y_predN))

```
